```python
# -----------------------------------------------
# EXO 5  EDA & Feature Engineering
# -----------------------------------------------

# Installer les bibliothques ncessaires
!pip install --quiet numpy pandas matplotlib seaborn

# Importer les bibliothques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset (dataset_part_2.csv depuis Exo 3)
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(URL)

# Afficher les 5 premires lignes
print("Preview of dataset_part_2.csv:")
print(df.head())

# --------------------------------------------------
# EDA: FlightNumber vs PayloadMass by Class
# --------------------------------------------------
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Payload Mass (kg)", fontsize=20)
plt.title("FlightNumber vs PayloadMass by Launch Outcome")
plt.show()

# --------------------------------------------------
# Extract Year from Date
# --------------------------------------------------
df['Year'] = df['Date'].str.split('-').str[0]
print("\nPreview of Year column added:")
print(df[['Date', 'Year']].head())

# --------------------------------------------------
# Feature Selection
# --------------------------------------------------
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 
               'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
print("\nSelected features:")
print(features.head())

# --------------------------------------------------
# One-Hot Encoding for Categorical Features
# --------------------------------------------------
categorical_columns = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']
features_one_hot = pd.get_dummies(features, columns=categorical_columns)
print("\nPreview of features after One-Hot Encoding:")
print(features_one_hot.head())

# --------------------------------------------------
# Cast all numeric columns to float64
# --------------------------------------------------
features_one_hot = features_one_hot.astype(float)
print("\nData types after casting:")
print(features_one_hot.dtypes)

# --------------------------------------------------
# Export the engineered dataset
# --------------------------------------------------
features_one_hot.to_csv('dataset_part_3.csv', index=False)
print("\nDataset exported to 'dataset_part_3.csv'")


# -----------------------------------------------
# EDA: Flight Number vs Launch Site
# -----------------------------------------------



# Scatter plot Flight Number vs Launch Site
plt.figure(figsize=(10,6))
sns.scatterplot(x='FlightNumber', y='LaunchSite', data=df, hue='Class', palette=['red','green'], s=100)
plt.title('Flight Number vs Launch Site')
plt.xlabel('Flight Number')
plt.ylabel('Launch Site')
plt.legend(title='Class', labels=['Failure', 'Success'])
plt.grid(True)
plt.show()

# Scatter plot: Payload Mass vs Launch Site
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='PayloadMass', 
    y='LaunchSite', 
    data=df, 
    hue='Class',       # 0 = failure, 1 = success
    palette=['red','green'], 
    s=100
)
plt.title('Payload Mass vs Launch Site')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Launch Site')
plt.legend(title='Class', labels=['Failure', 'Success'])
plt.grid(True)
plt.show()

# -----------------------------------------------
# Success Rate vs Orbit Type
# -----------------------------------------------


# Calculate success rate per orbit type
success_rate_orbit = df.groupby('Orbit')['Class'].mean().reset_index()

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(success_rate_orbit['Orbit'], success_rate_orbit['Class'], color='skyblue')
plt.title('Success Rate vs Orbit Type')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')
plt.ylim(0,1)
plt.grid(axis='y')
plt.show()

# Scatter plot Flight Number vs Orbit Type
plt.figure(figsize=(12,6))
sns.scatterplot(x='FlightNumber', y='Orbit', data=df, hue='Class', palette=['red','green'], s=100)
plt.title('Flight Number vs Orbit Type')
plt.xlabel('Flight Number')
plt.ylabel('Orbit Type')
plt.legend(title='Class', labels=['Failure', 'Success'])
plt.grid(True)
plt.show()

# Scatter plot Payload Mass vs Orbit Type
plt.figure(figsize=(12,6))
sns.scatterplot(x='PayloadMass', y='Orbit', data=df, hue='Class', palette=['red','green'], s=100)
plt.title('Payload Mass vs Orbit Type')
plt.xlabel('Payload Mass (kg)')
plt.ylabel('Orbit Type')
plt.legend(title='Class', labels=['Failure', 'Success'])
plt.grid(True)
plt.show()

# -----------------------------------------------
# Launch Success Yearly Trend
# -----------------------------------------------

# Assume df is already loaded with 'Date' and 'Class'
# Extract year from Date
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Calculate yearly average success rate
yearly_success = df.groupby('Year')['Class'].mean()

# Plot the trend
plt.figure(figsize=(10,6))
plt.plot(yearly_success.index, yearly_success.values, marker='o', linestyle='-')
plt.title('Launch Success Rate Over Years')
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.grid(True)
plt.show()

# Find unique launch sites
unique_launch_sites = df['LaunchSite'].unique()
print(unique_launch_sites)


# Filter for launch sites starting with 'CCA'
cca_sites = df[df['LaunchSite'].str.startswith('CCA')].head(5)
print(cca_sites)

# Extraire les launch sites commenant par 'CCA'
cca_sites_names = df['LaunchSite'][df['LaunchSite'].str.startswith('CCA')].unique()
print(cca_sites_names)

#-------------------------------#


# --- 1) Inspect columns (very important) ---
print("Columns in dataframe:")
print(df.columns.tolist())
print()

# --- 2) Ensure PayloadMass is numeric (coerce errors to NaN) ---
df['PayloadMass'] = pd.to_numeric(df['PayloadMass'], errors='coerce')

# --- 3) Try to filter by Customer if it exists, otherwise fallback to Orbit == 'LEO' ---
if 'Customer' in df.columns:
    # case-insensitive search for NASA inside Customer
    nasa_mask = df['Customer'].astype(str).str.contains('NASA', case=False, na=False)
    nasa_flights = df[nasa_mask]
    filter_used = "Customer contains 'NASA'"
elif 'Orbit' in df.columns:
    nasa_flights = df[df['Orbit'] == 'LEO']
    filter_used = "Orbit == 'LEO' (fallback)"
else:
    raise KeyError("Neither 'Customer' nor 'Orbit' columns found in dataframe; cannot filter NASA missions.")

# --- 4) Calculate totals, handling missing payloads ---
count_missions = len(nasa_flights)
total_payload_nasa = nasa_flights['PayloadMass'].sum(skipna=True)  # sum ignores NaN by default

print(f"Filter used: {filter_used}")
print(f"Number of missions selected: {count_missions}")
print(f"Total payload mass (kg) for selected missions: {total_payload_nasa:,.2f} kg")


# Filtrer les vols avec booster version "F9 v1.1"
f9_v11 = df[df['BoosterVersion'] == "F9 v1.1"]

# Calculer la masse moyenne du payload
average_payload = f9_v11['PayloadMass'].mean()

print("Average payload mass for F9 v1.1:", average_payload)
print(df['BoosterVersion'].unique())

# Filtrer les missions avec un atterrissage russi sur un pad au sol
ground_success = df[(df['Class'] == 1) & (df['LandingPad'].str.contains("LZ", na=False))]


# Afficher les valeurs uniques de LandingPad pour les succs
print(df[df['Class'] == 1]['LandingPad'].unique())
# Trouver la date la plus ancienne
first_ground_landing_date = ground_success['Date'].min()

print(first_ground_landing_date)

# Filtrer les atterrissages russis
success_landings = df[df['Class'] == 1]

# Trouver la date la plus ancienne
first_success_date = success_landings['Date'].min()
print(first_success_date)



print(df.columns)
# Filter boosters with successful drone ship landings and payload between 4000 and 6000
successful_ds_boosters = df[
    (df['LandingPad'] == 'OCISLY') &  
    (df['Outcome'] == 'Success') &
    (df['PayloadMass'] > 4000) &
    (df['PayloadMass'] < 6000)
]['BoosterVersion']

print(successful_ds_boosters.unique())


# Total number of successful and failed missions
mission_counts = df['Class'].value_counts()
print(mission_counts)

# Find the maximum payload mass
max_payload = df['PayloadMass'].max()

# List boosters which carried the maximum payload
boosters_max_payload = df[df['PayloadMass'] == max_payload]['BoosterVersion'].unique()
print(boosters_max_payload)

# Filter launches in 2015 with failed landings on drone ship (OCISLY)
failed_2015 = df[
    (df['Year'] == 2015) &
    (df['LandingPad'] == 'OCISLY') &
    (df['Outcome'] == 'Failure')
][['BoosterVersion', 'LaunchSite', 'Outcome']]

print(failed_2015)


# Filter launches between 2010-06-04 and 2017-03-20
filtered_df = df[
    (df['Date'] >= '2010-06-04') &
    (df['Date'] <= '2017-03-20')
]

# Count and rank landing outcomes
landing_outcome_counts = filtered_df['Outcome'].value_counts().sort_values(ascending=False)
print(landing_outcome_counts)
```
